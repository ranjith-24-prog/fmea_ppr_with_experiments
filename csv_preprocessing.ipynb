```python
import pandas as pd
import os
```

```python
# --- Step 1: Specify/input Excel filename ---
excel_input = "Excel_and_csv_files\Spot Welding FMEA form.xlsx"  # Change this to your uploaded Excel file's name
csv_output = "Excel_and_csv_files\converted_from_excel.csv"

# --- Step 2: Convert Excel to CSV ---
# This loads the first sheet (change sheet_name=... for others)
df_excel = pd.read_excel(excel_input, sheet_name=0)
df_excel.to_csv(csv_output, index=False)
print(f"Converted Excel ({excel_input}) to CSV: {csv_output}")
```

```python
csv_path = "Excel_and_csv_files\converted_from_excel.csv"

df = pd.read_csv(csv_path)
print("Loaded with inferred header. Shape:", df.shape)
display(df.head(10))
```

```python
# Drop first 5 rows (0..4) and reset index
df_trim = df.drop(index=list(range(0, 4)), errors='ignore').reset_index(drop=True)
print('\nAfter dropping first 5 rows (including any inferred header), shape:', df_trim.shape)
display(df_trim.head(8))
```

```python
# 2) Make the first row the header (promote row 0 to column names)
header = df_trim.iloc[0].astype(str).tolist()
df_trim = df_trim.iloc[1:].reset_index(drop=True)
df_trim.columns = header

print('\nAfter promoting first row to header, shape:', df_trim.shape)
print('Columns (first 10):', list(df_trim.columns)[:10])
display(df_trim.head(8))

# Insert a new first column named 'System element'
df_trim.insert(0, 'System element', None)

print("New shape after inserting first column:", df_trim.shape)
display(df_trim.head(10))
```

```python
# Drop columns where the column name is null/NaN/None or the literal string 'nan' (any case)
col_index = pd.Index(df_trim.columns)
cond_missing = col_index.isnull()
col_as_str = col_index.map(lambda x: str(x).strip() if not pd.isna(x) else x)
cond_literal_nan = col_as_str.map(lambda x: isinstance(x, str) and x.lower() == 'nan')
mask_invalid = cond_missing | cond_literal_nan
cols_to_drop = list(col_index[mask_invalid])

df_trim = df_trim.drop(columns=cols_to_drop)
print('Dropped columns with NaN-like headers:', cols_to_drop)
print('New shape:', df_trim.shape)
display(df_trim.head(8))
```

```python
import re

# Build a regex for markers like "System element: Weld Quality Control"
pattern = re.compile(r'^\s*System\s*element\s*:\s*(.*)\s*$', flags=re.IGNORECASE)

# Ensure text dtype for Function
func = df_trim['Function'].fillna('').astype(str)

# Extract the element names where markers appear
extracted = func.map(lambda x: pattern.match(x).group(1).strip() if pattern.match(x) else None)

# Optional normalization (e.g., title case)
extracted = extracted.map(lambda s: s.title() if isinstance(s, str) else s)

# Create the column if missing
if 'System element' not in df_trim.columns:
    df_trim.insert(0, 'System element', None)

# Write extracted names on marker rows, then forward-fill down
df_trim.loc[extracted.notna(), 'System element'] = extracted[extracted.notna()]
df_trim['System element'] = df_trim['System element'].ffill()

# Drop the marker rows from the data
marker_rows = func.map(lambda x: bool(pattern.match(x)))
df_trim = df_trim.loc[~marker_rows].reset_index(drop=True)

print('Applied System element grouping and removed marker rows.')
display(df_trim[['System element', 'Function']].head(20))
```

```python
# Fill forward all columns line-by-line to propagate previous non-null values downwards
df_trim = df_trim.fillna(method='ffill').reset_index(drop=True)

print("Performed forward fill on all columns to activate empty cells.")
display(df_trim.head(20))
```

```python
# Save the cleaned DataFrame to CSV
output_path = "Excel_and_csv_files\converted_csv_file_final.csv"
df_trim.to_csv(output_path, index=False)
print(f"Cleaned CSV saved as: {output_path}")
```
